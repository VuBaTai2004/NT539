```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Load data
df = pd.read_csv('./CICDDoS2019_clean_data_ver_30k_42seed.csv', index_col=None, header=0)

print(df.head(5))
print("Data info:")
print(df.info())
print("Label count:")
print(df['Label'].value_counts())

# Drop un-use columns
print("Drop un-use columns")
df = df.drop(columns = ['Unnamed: 0', 'Flow ID', ' Source IP', ' Destination IP',
                        ' Timestamp', 'SimillarHTTP'])
# Check null
print("Check Null for each Col")
print(df.isnull().sum()) # Corrected line

# Check infinity values
print("checking for infinity")
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

# df['Label'] = df['Label'].apply(lambda x: x if x == 'BENIGN' else "Attacks")
X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = X_train[:, -1]
y_test = X_test[:, -1]
X_train = X_train[:, :-1]
X_test = X_test[:, :-1]

print("Training data shape:", X_train.shape, y_train.shape)
print("Testing data shape:", X_test.shape, y_test.shape)

print("Label Train count:")
unique, counts = np.unique(y_train, return_counts=True)
print(unique, counts)

print("Label Test count:")
unique, counts = np.unique(y_test, return_counts=True)
print(unique, counts)

le = preprocessing.LabelEncoder().fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
print("Label name:", le.classes_)
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = MLPClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
cnf_matrix = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
tpr = recall_score(y_test, y_pred, average='weighted', zero_division = 0)
ppv = precision_score(y_test, y_pred, average='weighted', zero_division = 0)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division = 0)

print("Matrix:")
print(cnf_matrix)
print("Acc:")
print(acc)
print("TPR:")
print(tpr)
print("PPV:")
print(ppv)
print("F1:")
print(f1)
```

Decision Tree, KNN

```python
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
dt = tree.DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_dt_pred = dt.predict(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_knn_pred = knn.predict(X_test)
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_gnb_pred = gnb.predict(X_test)
```

```python
# cnf_matrix = confusion_matrix(y_test, y_pred)

acc_dt = accuracy_score(y_test, y_dt_pred)
tpr_dt = recall_score(y_test, y_dt_pred, average='weighted', zero_division = 0)
ppv_dt = precision_score(y_test, y_dt_pred, average='weighted', zero_division = 0)
f1_dt = f1_score(y_test, y_dt_pred, average='weighted', zero_division = 0)

acc_knn = accuracy_score(y_test, y_knn_pred)
tpr_knn = recall_score(y_test, y_knn_pred, average='weighted', zero_division = 0)
ppv_knn = precision_score(y_test, y_knn_pred, average='weighted', zero_division = 0)
f1_knn = f1_score(y_test, y_knn_pred, average='weighted', zero_division = 0)

acc_gnb = accuracy_score(y_test, y_gnb_pred)
tpr_gnb = recall_score(y_test, y_gnb_pred, average='weighted', zero_division = 0)
ppv_gnb = precision_score(y_test, y_gnb_pred, average='weighted', zero_division = 0)
f1_gnb = f1_score(y_test, y_gnb_pred, average='weighted', zero_division = 0)

# print(cnf_matrix)
```

```python
print("Decision Tree Metrics:")
print(f"Accuracy: {acc_dt}")
print(f"Recall (TPR): {tpr_dt}")
print(f"Precision (PPV): {ppv_dt}")
print(f"F1-Score: {f1_dt}")

print("\n\n")

print("KNN Metrics:")
print(f"Accuracy: {acc_knn}")
print(f"Recall (TPR): {tpr_knn}")
print(f"Precision (PPV): {ppv_knn}")
print(f"F1-Score: {f1_knn}")

print("\n\n")

print("Gaussian Naive Bayes Metrics:")
print(f"Accuracy: {acc_gnb}")
print(f"Recall (TPR): {tpr_gnb}")
print(f"Precision (PPV): {ppv_gnb}")
print(f"F1-Score: {f1_gnb}")

print("\n\n")

results = pd.DataFrame({
    'Model': ['Decision Tree', 'KNN', 'Gaussian Naive Bayes'],
    'Accuracy': [acc_dt, acc_knn, acc_gnb],
    'Recall': [tpr_dt, tpr_knn, tpr_gnb],
    'Precision': [ppv_dt, ppv_knn, ppv_gnb],
    'F1-Score': [f1_dt, f1_knn, f1_gnb]
})

print(results)
```
